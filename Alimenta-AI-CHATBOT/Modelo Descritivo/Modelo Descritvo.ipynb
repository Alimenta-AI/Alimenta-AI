### Explicação do agrupamento
Realizamos um agrupamento de estatísticas por país utilizando o DataFrame "dados". Calculamos a média e a soma do "Rank in 2022" para cada país. Essas estatísticas nos fornecem informações valiosas sobre a distribuição do "Rank in 2022" em diferentes países. O agrupamento nos permite analisar os dados de forma mais granular, identificando padrões e diferenças entre os países.
Espero que este relatório tenha sido útil para entender o agrupamento de estatísticas por país.

```python
grouped_data = dados.groupby('Country')  # Agrupa os dados por país

# Calcula a média para cada grupo
mean_by_country = grouped_data['Rank in 2022'].mean()
print("Média por país:")
print(mean_by_country)

# Calcula a soma para cada grupo
sum_by_country = grouped_data['Rank in 2022'].sum()
print("Soma por país:")
print(sum_by_country)

```
