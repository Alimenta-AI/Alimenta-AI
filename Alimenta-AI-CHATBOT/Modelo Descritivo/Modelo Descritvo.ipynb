### Explicao do agrupamento
Realizamos um agrupamento de estatsticas por pas utilizando o DataFrame "dados". Calculamos a mdia e a soma do "Rank in 2022" para cada pas. Essas estatsticas nos fornecem informaes valiosas sobre a distribuio do "Rank in 2022" em diferentes pases. O agrupamento nos permite analisar os dados de forma mais granular, identificando padres e diferenas entre os pases.
Espero que este relatrio tenha sido til para entender o agrupamento de estatsticas por pas.

```python
grouped_data = dados.groupby('Country')  # Agrupa os dados por pas

# Calcula a mdia para cada grupo
mean_by_country = grouped_data['Rank in 2022'].mean()
print("Mdia por pas:")
print(mean_by_country)

# Calcula a soma para cada grupo
sum_by_country = grouped_data['Rank in 2022'].sum()
print("Soma por pas:")
print(sum_by_country)

```
